First ,What is Customer Churn ?

### We will be using logistic Regression with different class_weights and different regularization rule to see which gives optimal results we will use our previous analysis to determine weights.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report ,confusion_matrix
```

```python
df["Churn"].value_counts()

```

```python
df['Gender'][df['Gender']=='Male']=1
df['Gender'][df['Gender']=='Female']=0
df['Subscription Type'][df['Subscription Type']=='Basic']=0
df['Subscription Type'][df['Subscription Type']=='Standard']=1
df['Subscription Type'][df['Subscription Type']=='Premium']=2
df['Contract Length'].replace(["Annual","Quarterly"    
,"Monthly"],[2,1,0    
],inplace=True)
df
```

```python
X=df.drop("Churn",axis=1)
weight_1=[0.01,0.1,0.2,0.25,0.40,0.50,0.70,0.2,0.56,0.5,0.3]
weight_2=[0.01,0.1,0.2,0.25,0.40,0.60,0.70,0.3,0.56,0.5,0.3]
x=X.values
min_max_scale=MinMaxScaler()
x=min_max_scale.fit_transform(x)
X=pd.DataFrame(x)
          
print(X)
print(len(weight_1))
```

```python
np.random.seed(42)
clf=LogisticRegression()
classweight_1=dict(zip([x for x in range(0,11)],weight_1))
classweight_2=dict(zip([x for x in range(0,11)],weight_2))
grid={"penalty":['l1','l2'],
       "class_weight":[classweight_1,classweight_2,"balanced"]}
clf_cv=GridSearchCV(clf,grid,cv=10,n_jobs=-1)
X_train,X_test,Y_train,Y_test=train_test_split(X,df["Churn"],test_size=0.8)

clf_cv.fit(X_train,Y_train)
```

```python
print(clf_cv.best_params_)
clf_cv.best_estimator_
```

```python
np.random.seed(42)
clf_2=LogisticRegression(class_weight='balanced')
clf_2.fit(X_train,Y_train)
clf_2.score(X_test,Y_test)
Y_preds=clf_2.predict(X_test)
table=classification_report(Y_test,Y_preds,output_dict=True)
print(confusion_matrix(Y_test,Y_preds))

pd.DataFrame(table)
```

```python
clf_cv.best_score_
```
