```python
df=pd.read_csv("emft- Sheet1.csv")
```

```python
df["Electric Field Intensity"]=np.zeros(13)
```

```python
df2=df.replace("Distribution type","Spherical")
df2["Distance"]=np.random.randint(20)
df2
```

```python
for dis in df2["Distribution type"]:
    if dis == "point charge":
        df2["Electric Field Intensity"]=9*10**(9)*df2["Charge "]/(df2["Distance"])**2
    if dis == "line charge":
        df2["Electric Field Intensity"]=2*9*10**(9)*df2["Charge "]/df2["Distance"]
    if dis == "Spherical":
        df2["Electric Field Intensity"]=2*9*10**(9)*df2["Charge "]/df2["Distance"]
          
        
```
