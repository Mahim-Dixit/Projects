# Creation of Series
1. From python objects
2. from numpy array

```python
#importing module
import pandas as pd
import numpy as np
```

```python
# list
l1=[4,5,6,8]
sr=pd.Series(l1)
sr
```

```python
#tuple
t1=(1,2,3,4,5)
sr=pd.Series(t1)
sr
```

```python
#dictionary
dic=dict(zip(t1,l1))
sr=pd.Series(dic)
sr
```
